```python
import pandas as pd
import random
from google.colab import widgets
import ipywidgets as widgets
```

```python
# data preprocess
df = pd.read_csv('Data.csv')
df = df[df['Suitable_Site_Conditions'].notna()]
df = df[df['Bloom_Time'].notna()]
df = df[df['Water_Needs'].notna()]
df = df[df['Plant_Type'].notna()]
df = df[df['Flower_Color'].notna()]
df = df[df['Size_at_Maturity'].notna()]
df = df[df['Common_Name'].notna()]

df = df[['Latin_Name','Common_Name','Family_Name','Plant_Type','Bloom_Time','Flower_Color','Size_at_Maturity','Climate_Appropriate_Plants','Suitable_Site_Conditions','Water_Needs','Appropriate_Location','Top20_int','tipes','photo']] # keep columns with less than 100 values missing
df = df[df.Plant_Type != 'Tree']
df = df[df.Plant_Type != 'Tree (deciduous)']
df = df[df.Plant_Type != 'Tree (evergreen)']
df['Top20_int'].fillna(0, inplace=True)
```

```python
# update the user new water needs
def Change_Water_Needs(df,previous_plant,changed_water_needs):
    previous_plant_water_needs = df[df['Common_Name'] == previous_plant]['Water_Needs'].tolist()[0]
    if changed_water_needs == 'can give more water':
        if previous_plant_water_needs == 'Low' or previous_plant_water_needs == 'Moderate':
            return 'Moderate'
        else:
            return 'Low'

    elif changed_water_needs == 'can give less water':
        if previous_plant_water_needs == 'None' or previous_plant_water_needs == 'Low':
            return 'None'
        else:
            return 'Low'

    elif changed_water_needs == "didn't change":
       return previous_plant_water_needs

    elif changed_water_needs == "not sure":
        return random.choice(['None','Low','Moderate'])
```

```python
# update the user new water needs
def Change_Site_Conditions(df,previous_plant,changed_site_conditions):
    previous_plant_suitable_site_conditions = df[df['Common_Name'] == previous_plant]['Suitable_Site_Conditions'].tolist()[0]
    if changed_site_conditions == 'more sun':
        if previous_plant_suitable_site_conditions == 'Part Shade' or previous_plant_suitable_site_conditions == 'Sun':
            return 'Sun'
        else:
            return 'Part Shade'


    elif changed_site_conditions == 'less sun':
        if previous_plant_suitable_site_conditions == 'Shade' or previous_plant_suitable_site_conditions == 'Part Shade':
            return 'Shade'
        else:
            return 'Part Shade'

    elif changed_site_conditions == "same":
        return previous_plant_suitable_site_conditions

    elif changed_site_conditions == "not sure":
        return random.choice(['Shade','Part Shade','Sun'])
```

```python
# update when the user will grow the plant
def Bloom_Time(bloom_time):
    if bloom_time == 'Yes':
           return 'Winter'

    elif bloom_time == 'next season':
            return 'Spring'

    elif bloom_time == "doesn't know/care":
        return random.choice(['Winter','Spring','Summer','Fall'])
```

```python
# gether previous plant from the user
previous_plant_dropdown = widgets.Dropdown(options=list(df['Common_Name']),description='Previous plant:',disabled=False)
previous_plant_dropdown
```

```python
# gether water routing changes from the user
changed_water_needs = widgets.Dropdown(options=["can give more water","can give less water","didn't change","not sure"],description='Water routine changed:',disabled=False)
changed_water_needs
```

```python
# gether site conditions changes from the user
changed_site_conditions = widgets.Dropdown(options=["more sun","less sun","same","not sure"],description='Place:',disabled=False)
changed_site_conditions
```

```python
# gether bloom time from the user
bloom_time = widgets.Dropdown(options=["yes","next season","doesn't know/care"],description='start now:',disabled=False)
bloom_time
```

##Run the next 3 cells to get your matching plant



```python
# create new point based extisting flower and new information from user
plant_dict = {}
previous_plant = previous_plant_dropdown.value
plant_dict['Latin_Name'] = df[df['Common_Name'] == previous_plant]['Latin_Name'].tolist()[0]
plant_dict['Family_Name'] = df[df['Common_Name'] == previous_plant]['Family_Name'].tolist()[0]
plant_dict['Water_Needs'] = Change_Water_Needs(df,previous_plant,changed_water_needs.value)
plant_dict['Suitable_Site_Conditions'] = Change_Site_Conditions(df,previous_plant,changed_site_conditions.value)
plant_dict['Bloom_Time'] = Bloom_Time(bloom_time.value)
plant_dict['Plant_Type'] = df[df['Common_Name'] == previous_plant]['Plant_Type'].tolist()[0]
plant_dict['Flower_Color'] = df[df['Common_Name'] == previous_plant]['Flower_Color'].tolist()[0]
plant_dict['Size_at_Maturity'] = df[df['Common_Name'] == previous_plant]['Size_at_Maturity'].tolist()[0]
plant_dict['Climate_Appropriate_Plants'] = df[df['Common_Name'] == previous_plant]['Climate_Appropriate_Plants'].tolist()[0]
plant_dict['Appropriate_Location'] = df[df['Common_Name'] == previous_plant]['Appropriate_Location'].tolist()[0]
plant_dict['Common_Name'] = 'name'
plant_dict['Top20_int'] = df[df['Common_Name'] == previous_plant]['Top20_int'].tolist()[0]
```

```python
# create df with all binary features for our dataset
plant_df = pd.DataFrame([plant_dict])
df = pd.concat([df, plant_df], ignore_index=True)

df['Suitable_Site_Conditions'] = df['Suitable_Site_Conditions'].str.split(';')
new_df = df.explode('Suitable_Site_Conditions')

df['Flower_Color'] = df['Flower_Color'].str.split(';')
new_df = new_df.explode('Flower_Color')

new_df['Appropriate_Location'] = df['Appropriate_Location'].str.split(';')
new_df = new_df.explode('Appropriate_Location')

new_df['Bloom_Time'] = new_df['Bloom_Time'].str.split(';')
new_df = new_df.explode('Bloom_Time')

full_df = pd.get_dummies(new_df, columns = ['Family_Name','Suitable_Site_Conditions','Flower_Color','Appropriate_Location', 'Bloom_Time', 'Water_Needs','Size_at_Maturity','Plant_Type','Climate_Appropriate_Plants'])
full_df = full_df.dropna()
full_df = full_df[full_df.Common_Name != previous_plant]
```

```python
# create arrays for KNN and implement 
array_no_names = full_df.drop(columns = ['Latin_Name','Common_Name','Top20_int','tipes','photo']).to_numpy()
common_names = full_df['Common_Name'].to_numpy()
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=1)
neigh.fit(array_no_names[:-1], common_names[:-1])
x = array_no_names[-1]
match_plant = str(neigh.predict([x])[0])

# show user his match plant
print("Your match plant is : " +  match_plant)
tip = df['tipes'][df['Common_Name'] ==  match_plant]
print("here is a tip from us : " + tip.values[0])
photo = df['photo'][df['Common_Name'] ==  match_plant]
print("you can see photo here: " , photo.values[0])
    
```
